```python
import random
import time
import pandas as pd

# Simulate IoT sensor data (temperature readings)
def generate_sensor_data(n=20):
    return [round(random.uniform(20.0, 40.0), 2) for _ in range(n)]

# Edge processing function: filter out normal data
def edge_filter(data, threshold=30.0):
    print(f"\n[Edge] Processing {len(data)} sensor readings...")
    filtered = [temp for temp in data if temp > threshold]
    print(f"[Edge] {len(filtered)} anomalous readings found (>{threshold}°C)")
    return filtered

# Simulate cloud receiving only filtered data
def send_to_cloud(filtered_data):
    print(f"\n[Cloud] Receiving {len(filtered_data)} readings...")
    df = pd.DataFrame(filtered_data, columns=["Temperature"])
    print("[Cloud] Data Received:")
    display(df)

# Run simulation
sensor_data = generate_sensor_data()
anomalous_data = edge_filter(sensor_data, threshold=30.0)
send_to_cloud(anomalous_data)
```
